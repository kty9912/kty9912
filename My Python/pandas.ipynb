```python
corona_df_2 = corona_df.sort_values("등록일시")
corona_df_2
```

```python
corona_df_2["일일 확진자"]= (corona_df_2["확진자"] 
                        - corona_df_2["확진자"].shift()).fillna(0)
# .fillna 결측값 NaN을 0으로 바꿔줌
corona_df_2["일일 사망자"]= (corona_df_2["사망자"].diff()).fillna(0)

corona_df_2
```

```python
corona_df_2["일일 사망자"].value_counts(ascending=True)
```

```python
corona_df_3 = corona_df_2.drop(["인덱스", "기준일", "게시글번호", 
                                "기준시간", "수정일시"], axis=1)
corona_df_3
```

```python
corona_df_3.reset_index(drop=True, inplace=True)
# inplace 원문 수정 (인덱스)
corona_df_3
```
