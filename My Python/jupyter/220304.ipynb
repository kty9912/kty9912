```python
import pandas as pd
```

```python
sales_df =pd.read_csv("./csv/Sales Records.csv")
sales_df.head()
```

```python
## 지금부터 해야될것
# 1. 컬럼삭제 ( ship date, order id, sales channel, Total revenue) --> 원본데이터 유지 (새로운 변수)
# 2. 컬럼명 수정 --> 원본데이터 유지 X (새로운 변수를 유지할 필요 없이 계속 같은 변수에서 작업 2번 ~ 4번) 
# ["지역", "국가", "상품타입", "배송순위", "주문날짜", "판매개수", "단가", "원가", "총비용", "총이윤"] 
# 3. Country 기준으로 sort
# 4. index 값을 초기화
```

```python
sales_df_2 = sales_df.drop(["Ship Date", "Order ID", "Sales Channel", "Total Revenue"], axis=1)
sales_df_2.head()
```

```python
sales_df_2.columns = ["지역", "국가", "상품타입", "배송순위", "주문날짜", 
                    "판매개수", "단가", "원가", "총비용", "총이윤"]
sales_df_2.head()
```

```python
sales_df_2.sort_values("국가", inplace=True)
sales_df_2
```

```python
sales_df_2.reset_index(drop=True, inplace=True)
sales_df_2
```

```python
sales_df_2["총이윤"] = sales_df_2["총이윤"].astype(int)
```

```python
sales_df_2.head()
```

```python
sales_df_2["총비용"] = sales_df_2["총비용"].round(0).astype(int)
sales_df_2.head()
```

```python
sales_df_2["원가"] = sales_df_2["원가"].round(1)
sales_df_2.head()
```

```python
sales_df_2["판매개수"].round(-1)
```

```python
sales_df_2["주문날짜"] = pd.to_datetime(sales_df_2["주문날짜"], 
                        format = "%m/%d/%Y")
```

```python
sales_df_2.head()
```

```python
sales_df_2.groupby(["지역", "국가", "상품타입"]).sum()["총이윤"]
# 평균은 mean()
```

```python
sales_df_2.groupby("국가").get_group("South Korea")
```

```python
sales_df_3 = sales_df_2[sales_df_2['국가'].isin(["South Korea", "China", "Togo"])]
sales_df_3.groupby('국가').mean()['총이윤']
```
